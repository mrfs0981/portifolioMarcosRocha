```python
import folium
import pandas as pd
from folium import plugins
from folium.plugins import MeasureControl
import xlrd
import xlwt
```

```python
df = pd.read_excel('trafego.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
coordenadas=[]
for lat,lng in zip(df.geolocation_lat,df.geolocation_lng):
    coordenadas.append([lat,lng])
```

```python
mapa = folium.Map(location=[-15.788497,-47.879873],zoom_start=5,tiles='Stamen Terrain')
```

```python
mapa.add_child(plugins.HeatMap(coordenadas))
mapa.add_child(folium.LatLngPopup())
mapa.add_child(MeasureControl())
mapa
```

```python
mapa.save('vivo.html')
```
